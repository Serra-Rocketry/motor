### Cálculo do Nozzle 

```python
import numpy as np
```

1. **Escolher o empuxo e a pressão na câmara de combustão do motor e inicialização de variaveis**  
   Definir os valores desejados de empuxo $(F)$ e pressão na câmara $(P_c)$.


```python
K = 1.042                   
F = 800                     # Empuxo 
# c_v = 1                   # Coeficiente de descarga do bocal (adimensional)
# c_p = 1.042               # Coeficiente de capacidade térmica a pressão constante (adimensional)
R = 208.37                  # Constante dos gases 
T_0 = 1600                  # Temperatura de combustão
P_0 = 68                    # Pressão da câmara de combustão em atm 
P_e = 1                     # Pressão ambiente 
C = 0                       # Característica do escoamento 
A_b = 0                     # Área de queima do grão
angulo_divergente = 15      # Ângulo divergente do nozzle em graus
angulo_convergente = 40     # Ângulo convergente do nozzle em grausm graus
diametro_camara = 0.040     # Diâmetro da câmara de combustão em mm
```

2. **Conhecer as propriedades termodinâmicas do fluido**  
   Coletar os parâmetros termodinâmicos:  
   - Razão de calores específicos: $$ k = \frac{c_p}{c_v} $$  

```python
# k = C_p/c_v
print("k = ", K)
```

3. **Calcular a razão de pressão para gerar escoamento engasgado (choked flow)**  
   Determinar a relação crítica de pressão para fluxo sônico na garganta:  
   $$
   \frac{P^*}{P_0} = \left( \frac{2}{k + 1} \right)^{\frac{k}{k - 1}}
   $$  
   onde $(P^*)$ é a pressão na garganta.


```python
R_pressao = np.pow((2/(K+1)), (K/(K-1)))

print(f'R_pressao: {R_pressao}\t')
```

4. **Calcular a razão entre a área da garganta e a área de saída**  
   Utilizar a relação de áreas para escoamento compressível:  
   $$
   Rarea = \frac{A^*}{A_e} = \left( \frac{k + 1}{2} \right)^{\frac{1}{k - 1}} \left( \frac{P_e}{P_0} \right)^{\frac{1}{k}}  \sqrt{\frac{k + 1}{k - 1} \left( 1 - \left( \frac{P_e}{P_0} \right)^{\frac{k - 1}{k}} \right)}
   $$  
   onde $(A^*)$ é a área da garganta e $(A_e)$ a área de saída.

```python
P_ratio = (P_e/P_0)

R_area = ((np.pow(((K+1)/2), (1/(K-1))))*(np.pow(P_ratio, (1/K))))*(np.sqrt(((K+1)/(K-1))*(1-(np.pow(P_ratio, (K-1)/K)))))
print(f'Resultado: {R_area}\t')
```

5. **Calcular a velocidade de saída $(v_e)$**  
   Aplicar a equação de velocidade isentrópica:  
   $$
   v_e = \sqrt{\frac{2k}{k - 1} R T_0 \left(  \left( 1 -\frac{P_e}{P_0} \right)^{\frac{k - 1}{k}} \right)}
   $$
   - Constante universal dos gáses (k/massa molar(fluido)): $$ R $$  
   - Temperatura total na câmara: $$ T_0 $$

```python
v_e = np.sqrt(((2*K)/(K - 1))*R*T_0*(np.pow(1-P_ratio, ((K - 1)/K))))
print(f'Velocidade de saída (v_e): {v_e}')
```

6. **Determinar a área da garganta $(A^*)$ e a área de saída**  
     $$
     A^* = \frac{\dot{m}}{P_0} \ {C^*}
     $$  

```python
m = F / v_e  # Vazão mássica (kg/s)
print(f'Vazão mássica (m): {m} kg/s')
C = np.sqrt(((R * T_0) / (K*(np.pow((2/(K+1)),((K+1)/(K-1)))))))  # Adicione após o cálculo de K
print(f'Constante C: {C}')

Area_garganta = (m/(P_0*101325))*C

A_e = Area_garganta / R_area

print(f'Área da garganta (A): {Area_garganta:.4f} m^2')
print(f'Área de saída (A_e): {A_e} m^2')
```

7. **Escolher uma configuração de grão compatível**  
   Garantir que a geometria do grão $\ ( \ A_b \ )$ satisfaça:  
   $$
   K_n = \frac{\ A_b}{A^*}
   $$  


```python
K_n = A_b/Area_garganta
print(f'Coeficiente de descarga (K_n): {K_n}')
```

8. **Determinar o comprimento do nozzle através dos ângulos de convergência/divergência**  
  

```python
diametro_garganta = 2*(np.sqrt(Area_garganta/np.pi))
print(f'Diâmetro da garganta: {diametro_garganta} m')
diametro_saida = 2 * (np.sqrt(A_e/np.pi))
print(f'Diâmetro de saída: {diametro_saida} m')

comprimento_divergente = ((diametro_saida - diametro_garganta)/(2*(np.tan(np.radians(angulo_divergente)))))
print(f'Comprimento divergente: {comprimento_divergente} m')
comprimento_convergente = (diametro_camara- diametro_garganta)/(2*np.tan(np.radians(angulo_convergente)))
print(f'Comprimento convergente: {comprimento_convergente} m')
```
